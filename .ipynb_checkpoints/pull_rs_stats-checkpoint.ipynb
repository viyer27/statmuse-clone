```python
import pandas as pd
import time
import random
import os

# Your original function (unchanged)
def get_per_game_stats(player_id, name):
    url = f"https://www.basketball-reference.com/players/{player_id[0]}/{player_id}.html"
    try:
        tables = pd.read_html(url)
        per_game_df = None

        for df in tables:
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = df.columns.get_level_values(1)
            if 'Season' in df.columns and 'PTS' in df.columns:
                per_game_df = df
                break

        if per_game_df is None:
            print(f"{name} ❌ Table not found")
            return None

        per_game_df = per_game_df.copy()
        per_game_df['Season'] = per_game_df['Season'].astype(str)
        per_game_df['Team'] = per_game_df['Team'].astype(str)
        per_game_df.dropna(how='all', inplace=True)
        per_game_df = per_game_df[~per_game_df['Season'].str.contains(r'\(\d+ Yrs\)', na=False)].copy()

        non_summary = per_game_df[~per_game_df['Season'].str.contains('Career|Yrs', na=False)].copy()
        duplicate_seasons = non_summary[non_summary.duplicated('Season', keep=False)]

        rows_to_drop = []
        for season in duplicate_seasons['Season'].unique():
            season_rows = per_game_df[per_game_df['Season'] == season]
            if any(season_rows['Team'].str.contains(r'TOT|\d+TM', na=False)):
                drop_rows = season_rows[~season_rows['Team'].str.contains(r'TOT|\d+TM', na=False)].index
                rows_to_drop.extend(drop_rows)

        per_game_df = per_game_df.drop(index=rows_to_drop)

        for col in per_game_df.columns:
            if col not in ['Season', 'Team', 'Lg', 'Pos', 'Awards']:
                per_game_df.loc[:, col] = pd.to_numeric(per_game_df[col], errors='coerce')

        per_game_df.insert(0, 'player_id', player_id)
        per_game_df.insert(1, 'Name', name)
        per_game_df.reset_index(drop=True, inplace=True)
        print(f"{name} ✅")
        return per_game_df

    except Exception as e:
        print(f"{name} ❌ Failed: {e}")
        return None

# Replace with your full player DataFrame
players_df = df
csv_output_path = "per_game_stats_progress.csv"

# Create CSV file if it doesn’t exist
if not os.path.exists(csv_output_path):
    pd.DataFrame().to_csv(csv_output_path, index=False)

batch_results = []
for idx, row in players_df.iterrows():
    df = get_per_game_stats(row['player_id'], row['Name'])
    if df is not None:
        batch_results.append(df)

    # Save every 500 players
    if (idx + 1) % 500 == 0 or (idx + 1) == len(players_df):
        print(f"Saving batch ending at player #{idx + 1}...")
        pd.concat(batch_results, ignore_index=True).to_csv(csv_output_path, mode='a', header=not os.path.getsize(csv_output_path), index=False)
        batch_results = []  # clear after saving

    # Gentle delay
    time.sleep(random.uniform(1.2, 2.0))
```

```python
import pandas as pd
import unicodedata
import re

def normalize_name(name):
    # Remove accents/special characters
    name = unicodedata.normalize('NFKD', name)
    name = name.encode('ASCII', 'ignore').decode('utf-8')
    return name

def generate_bbref_id(name):
    name = normalize_name(name)
    parts = name.lower().split()
    if len(parts) < 2:
        return None  # Not enough info

    first, last = parts[0], parts[-1]
    last_part = re.sub(r'[^a-z]', '', last)[:5].ljust(5, 'x')
    first_part = re.sub(r'[^a-z]', '', first)[:2].ljust(2, 'x')
    return last_part + first_part + "01"

# Fix player_id
players_df['corrected_player_id'] = players_df['Name'].apply(generate_bbref_id)

players_df['player_id'] = players_df['corrected_player_id']
players_df.drop(columns='corrected_player_id', inplace=True)
players_df.head()
```

```python
import pandas as pd

# Get all columns starting after 'Name'
cols_after_name = df1.columns[df1.columns.get_loc('Name') + 1:]

# Drop rows where all columns after 'Name' are NaN
df1 = df1[~df1[cols_after_name].isna().all(axis=1)].reset_index(drop=True)
df1 = df1[~df1['Season'].astype(str).str.contains('Yr', na=False)].reset_index(drop=True)
```

```python
def parse_awards(cell):
    result = {
        'MVP': False,
        'ROY': False,
        'All_Star': False,
        'All_NBA_1st': False,
        'All_NBA_2nd': False,
        'All_NBA_3rd': False,
        'All_Def_1st': False,
        'All_Def_2nd': False,
        'MIP': False,
        'DPOY': False,
        'Sixth_Man': False,
        'Finals_MVP': False,
        'AllStar_MVP': False
    }

    if pd.isna(cell):
        return pd.Series(result)
    
    parts = cell.split(',')
    for part in parts:
        if part == 'AS':
            result['All_Star'] = True
        elif part == 'MVP-1':
            result['MVP'] = True
        elif part == 'ROY-1':
            result['ROY'] = True
        elif part == 'MIP-1':
            result['MIP'] = True
        elif part == 'DPOY-1':
            result['DPOY'] = True
        elif part == '6MOY-1':
            result['Sixth_Man'] = True
        elif part == 'FMVP-1':
            result['Finals_MVP'] = True
        elif part == 'SMVP-1':
            result['AllStar_MVP'] = True
        elif part == 'NBA1':
            result['All_NBA_1st'] = True
        elif part == 'NBA2':
            result['All_NBA_2nd'] = True
        elif part == 'NBA3':
            result['All_NBA_3rd'] = True
        elif part == 'DEF1':
            result['All_Def_1st'] = True
        elif part == 'DEF2':
            result['All_Def_2nd'] = True
    
    return pd.Series(result)


# Apply to 'Awards' column
award_cols = df1['Awards'].apply(parse_awards)
df1 = pd.concat([df1, award_cols], axis=1)
```
